```python
import os
import pandas as pd

# Specify the directory where the files are stored
folder_path = 'DT Level'

# List all files in the directory
files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Initialize an empty dataframe to store all data
combined_df = pd.DataFrame()

# Loop through the files and append them to the combined dataframe
for file in files:
    # Construct full file path
    file_path = os.path.join(folder_path, file)
    # Read the Excel file
    data = pd.read_excel(file_path)
    # Append the data to the combined dataframe
    combined_df = pd.concat([combined_df, data], ignore_index=True)

# Display the combined dataframe 
combined_df.head()
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

def analyze_bias_and_stereotypes(df):
    # Define keywords related to biases and stereotypes
    bias_keywords = {
        'gender': ['woman', 'man', 'female', 'male', 'girl', 'boy'],
        'race': ['black', 'white', 'asian', 'latino'],
        'age': ['old', 'young', 'teenager', 'elderly'],
        'religion': ['christian', 'muslim', 'jew', 'hindu', 'atheist'],
        'occupation': ['doctor', 'engineer', 'teacher', 'nurse'],
        'lgbtq': ['gay', 'lesbian', 'bisexual', 'transgender'],
        'other': ['fat', 'skinny', 'smart', 'stupid']
    }
    
    # Initialize columns for bias keyword count and sentiment
    for category in bias_keywords:
        df[f'{category}_bias_count'] = 0
    
    df['sentiment'] = 0.0
    df['emotion'] = None

    # Analyze dialogues for bias keywords and sentiment
    for index, row in df.iterrows():
        dialogue = row['dialogue'].lower()
        sentiment_score = sia.polarity_scores(dialogue)
        
        # Update sentiment column with compound sentiment score
        df.at[index, 'sentiment'] = sentiment_score['compound']
        
        # Determine the predominant emotion
        if sentiment_score['compound'] >= 0.05:
            df.at[index, 'emotion'] = 'positive'
        elif sentiment_score['compound'] <= -0.05:
            df.at[index, 'emotion'] = 'negative'
        else:
            df.at[index, 'emotion'] = 'neutral'

        # Count occurrences of bias keywords
        for category, keywords in bias_keywords.items():
            count = sum(dialogue.count(keyword) for keyword in keywords)
            df.at[index, f'{category}_bias_count'] = count

    # Summarize bias counts across different movies
    bias_summary = df.groupby('movie').agg({f'{cat}_bias_count': 'sum' for cat in bias_keywords})
    bias_summary['total_bias_count'] = bias_summary.sum(axis=1)
    
    return df, bias_summary

# Assuming 'combined_df' is your DataFrame
combined_df, bias_summary = analyze_bias_and_stereotypes(combined_df)

# Display the first few rows of the modified DataFrame and the summary
print(combined_df.head())
print(bias_summary.head())
```

```python

# Define the boundary year to separate the movies into two groups
boundary_year = 2000

# Create two subsets of the data based on the year
movies_1900s = combined_df[combined_df['year'] < boundary_year]
movies_2000s = combined_df[combined_df['year'] >= boundary_year]
```

```python
# Define a function to calculate bias and sentiment metrics
def calculate_metrics(df):
    # Aggregate bias counts
    bias_columns = [col for col in df.columns if 'bias_count' in col]
    bias_summary = df[bias_columns].sum().to_frame(name='count').transpose()
    
    # Calculate average sentiment
    average_sentiment = df['sentiment'].mean()
    bias_summary['average_sentiment'] = average_sentiment
    
    # Count sentiment categories
    sentiment_counts = df['emotion'].value_counts(normalize=True)
    bias_summary = pd.concat([bias_summary, sentiment_counts], axis=1)
    
    return bias_summary

# Apply the function to both subsets
metrics_1900s = calculate_metrics(movies_1900s)
metrics_2000s = calculate_metrics(movies_2000s)
```

```python
# Print the metrics to compare them directly
print("Bias and Sentiment Metrics for Movies in the 1900s:")
print(metrics_1900s)
print("\nBias and Sentiment Metrics for Movies in the 2000s:")
print(metrics_2000s)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'combined_df' is your DataFrame and it contains a 'year' and bias columns
# Create a new column to classify movies by century
combined_df['century'] = combined_df['year'].apply(lambda x: '1900s' if x < 2000 else '2000s')

# Bias categories identified earlier
bias_columns = [col for col in combined_df.columns if 'bias_count' in col]

# Group data by movie and century for analysis
grouped_df = combined_df.groupby(['movie', 'century'])[bias_columns].sum().reset_index()
```

```python
# Filter data for each century
movies_1900s = grouped_df[grouped_df['century'] == '1900s']
movies_2000s = grouped_df[grouped_df['century'] == '2000s']

# Function to plot biases for a given DataFrame
def plot_biases(df, century):
    num_plots = len(bias_columns)
    cols = 2
    rows = num_plots // cols + (num_plots % cols > 0)
    
    fig, axes = plt.subplots(rows, cols, figsize=(14, rows * 4))
    axes = axes.flatten()
    
    for idx, bias in enumerate(bias_columns):
        ax = axes[idx]
        sns.barplot(data=df, x='movie', y=bias, ax=ax)
        ax.set_title(f'{bias} in {century}')
        ax.set_ylabel('Counts')
        ax.set_xlabel('Movies')
        ax.tick_params(axis='x', rotation=90)
    
    plt.tight_layout()
    plt.show()

# Plot for 1900s
plot_biases(movies_1900s, '1900s')

# Plot for 2000s
plot_biases(movies_2000s, '2000s')
```

```python
# Aggregate data by century for heatmap comparison
century_comparison = combined_df.groupby(['century'])[bias_columns].sum()

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(century_comparison, annot=True, cmap='viridis', fmt='g')
plt.title('Comparative Heatmap of Bias Counts by Century')
plt.ylabel('Century')
plt.xlabel('Bias Types')
plt.show()
```
